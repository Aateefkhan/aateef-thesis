```python
import pandas as pns
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.arima_model import ARIMA
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
disk = pns.read_csv('/root/aateef-thesis/jupyter/csv/disk.csv', parse_dates = [0], index_col = [0])
```

```python
disk
```

```python
plot_acf(disk, lags=3)
```

```python
plot_pacf(disk, lags=3)
```

```python
disk.describe()
disk_mod = ARIMA(disk, order=(1,1,0))
```

```python
disk_mod_fit = disk_mod.fit()
```

```python
disk_mod_fit.aic
```

```python
disk_forecast = disk_mod_fit.forecast(steps = 3)[0]
```

```python
disk_forecast
```
